```python
try:
    SCROLL_PAUSE_TIME = 3.5
    driver.get("https://www.documentonews.gr/tag/%ce%b4%ce%b7%ce%bc%ce%b7%cf%84%cf%81%ce%b7%cf%83-%ce%ba%ce%bf%cf%85%cf%86%ce%bf%ce%bd%cf%84%ce%b9%ce%bd%ce%b1%cf%83/")


    last_height = driver.execute_script("return document.body.scrollHeight")

    while (last_height < 69000):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        new_height = driver.execute_script("return document.body.scrollHeight")
        print('height:', new_height)
        if new_height == last_height:
            print('height =' , new_height)
            break
        last_height = new_height



finally:
#    driver.quit()
    print("OK")
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
stories_list = []

stories = soup.find_all('div', { 'class': 'item post loop' })

for story in stories:
    heading = story.find('h2')
    link = heading.find('a')
    date = story.find('div', {'class': 'item-meta'})
    # datetime = date['datetime']
    summary = story.find('div', {'class': 'entry-summary'})


    
    story_dict = {
        'headline': heading.text,
        'url': link['href'],
        'date': date,
        'summary': summary

    }
     
    # Πρόσθεσε το λέξικο στη λίστα
    stories_list.append(story_dict)
```

```python
print(len(stories_list))
```

```python

import pandas as pd
df = pd.DataFrame(stories_list)
```
